```python
def webcam():
    cam = cv2.VideoCapture(0)
    cv2.namedWindow("test")

    img_counter = 0
    dis=''
    while True:
        ret, frame = cam.read()
        cv2.imshow("test", frame)
        if not ret:
            break
        k = cv2.waitKey(1)

        if k%256 == 27:
            # ESC pressed
            print("Escape hit, closing...")
            engine = pyttsx3.init()
            engine.say(dis)
            engine.runAndWait()
            break
             
        elif k%256 == 32:
            # SPACE pressed
            img_name = "opencv_frame_0.jpg"
            cv2.imwrite(img_name, frame)
            print("{} written!".format(img_name))
            
            op=cv2.imread('./opencv_frame_0.jpg')
            op=cv2.cvtColor(op,cv2.COLOR_BGR2GRAY)
            op1=skimage.transform.resize(op,(50,50,1))
            img_arr=np.asarray(op1)
            op2=[]
            op2.append(img_arr)
            op2=np.asarray(op2)
            op3=n.predict(op2)
            z=-1
            i1=-1
            for i in range(0,30):
                if z<op3[0][i]:
                    z=op3[0][i]
                    i1=i
            if i1==0:
                dis=dis+'A'
                T.insert(tk.END, 'A')
                T.update()
            elif i1==1:
                dis=dis+'B'
                T.insert(tk.END, 'B')
                T.update()
            elif i1==2:
                dis=dis+'C'
                T.insert(tk.END, 'C')
                T.update()
            elif i1==3:
                dis=dis+'D'
                T.insert(tk.END, 'D')
                T.update()
            elif i1==4:
                dis=dis+'E'
                T.insert(tk.END, 'E')
                T.update()
            elif i1==5:
                dis=dis+'F'
                T.insert(tk.END, 'F')
                T.update()
            elif i1==6:
                dis=dis+'G'
                T.insert(tk.END, 'G')
                T.update()
            elif i1==7:
                dis=dis+'H'
                T.insert(tk.END, 'H')
                T.update()
            elif i1==8:
                dis=dis+'I'
                T.insert(tk.END, 'I')
                T.update()
            elif i1==9:
                dis=dis+'J'
                T.insert(tk.END, 'J')
                T.update()
            elif i1==10:
                dis=dis+'K'
                T.insert(tk.END, 'K')
                T.update()
            elif i1==11:
                dis=dis+'L'
                T.insert(tk.END, 'L')
                T.update()
            elif i1==12:
                dis=dis+'M'
                T.insert(tk.END, 'M')
                T.update()
            elif i1==13:
                dis=dis+'N'
                T.insert(tk.END, 'N')
                T.update()
            elif i1==14:
                dis=dis+'O'
                T.insert(tk.END, 'O')
                T.update()
            elif i1==15:
                dis=dis+'P'
                T.insert(tk.END, 'P')
                T.update()
            elif i1==16:
                dis=dis+'Q'
                T.insert(tk.END, 'Q')
                T.update()
            elif i1==17:
                dis=dis+'R'
                T.insert(tk.END, 'R')
                T.update()
            elif i1==18:
                dis=dis+'S'
                T.insert(tk.END, 'S')
                T.update()
            elif i1==19:
                dis=dis+'T'
                T.insert(tk.END, 'T')
                T.update()
            elif i1==20:
                dis=dis+'U'
                T.insert(tk.END, 'U')
                T.update()
            elif i1==21:
                dis=dis+'V'
                T.insert(tk.END, 'V')
                T.update()
            elif i1==22:
                dis=dis+'W'
                T.insert(tk.END, 'W')
                T.update()
            elif i1==23:
                dis=dis+'X'
                T.insert(tk.END, 'X')
                T.update()
            elif i1==24:
                dis=dis+'Y'
                T.insert(tk.END, 'Y')
                T.update()
            elif i1==25:
                dis=dis+'Z'
                T.insert(tk.END, 'Z')
            elif i1==26:
                if dis=='':
                    continue
                dis=dis[0:-1]
                T.delete(1.0,tk.END)
                T.update()
                T.insert(tk.END,dis)
                T.update()
            elif i1==27:
                T.insert(tk.END, '')
                T.update()
            elif i1==28:
                dis=dis+' '
                T.insert(tk.END, ' ')
                T.update()
            elif i1==29:
                T.insert(tk.END, '') 

            os.remove("opencv_frame_0.jpg")
            img_counter += 1
            temp=dis
    
    cam.release()
    T.delete(1.0,tk.END)
    T.update()
    cv2.destroyAllWindows()
```
